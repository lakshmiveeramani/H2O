```python
import h2o
```

```python
import h2o.grid
```

```python
h2o.init()
```

```python
data = h2o.import_file("http://h2o-public-test-data.s3.amazonaws.com/smalldata/airlines/allyears2k_headers.zip")
```

```python
train, valid, test = data.split_frame([0.8,0.1], seed = 69)
```

```python
print("%d/%d/%d"  % (train.nrows, valid.nrows, test.nrows))
```

```python
y = "IsArrDelayed"
```

```python
ignoreFields = ["ArrDelay", "DepDelay",
                  "CarrierDelay", "WeatherDelay",
                  "NASDelay", "SecurityDelay",
                  "IsDepDelayed", "IsArrDelayed",
                  "ActualElapsedTime"]
```

```python
xAll = [ i for i in train.names if i not in ignoreFields]
```

```python
xLikely = ["Month", "DayOfWeek", "UniqueCarrier",
             "Origin", "Dest", "Distance",
             "Cancelled", "Diverted"]
```

```python
from h2o.estimators.glm import H2OGeneralizedLinearEstimator
```

```python
m_def = H2OGeneralizedLinearEstimator(family = "binomial")
```

```python
m_def.train(xAll, y, train, validation_frame = valid)
```

```python
m_def.logloss(valid = True)
```

```python
g = h2o.grid.H2OGridSearch(
H2OGeneralizedLinearEstimator(family = "binomial",
                             lambda_search = True
                          ),
                          hyper_params = {
                              "alpha" : [x*0.01 for x in range(0,100)],
                          },
    
                          search_criteria = {
                              "strategy":"RandomDiscrete",
                              "max_models":8,
                              "max_runtime_secs":30
                          }
)
        
```

```python
g.train(xAll, y, train, validation_frame = valid)
```

```python
g
```

```python
g = h2o.grid.H2OGridSearch(
H2OGeneralizedLinearEstimator(family = "binomial",
                             lambda_search = True
                          ),
                          hyper_params = {
                              "alpha" : [0, 0.2, 0.4, 0.5, 0.6, 0.8, 0.99],
                          },
    
                          search_criteria = {
                              "strategy":"Cartesian",

                          }
)
g.train(xLikely, y, train, validation_frame = valid)
```
