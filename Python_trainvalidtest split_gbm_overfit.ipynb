```python
train = h2o.get_frame("people_train")
```

```python
valid = h2o.get_frame("people_valid")
```

```python
test = h2o.get_frame("people_test")
```

```python
from h2o.estimators.gbm import H2OGradientBoostingEstimator
```

```python
y = "income"
```

```python
ignoreFields = [y, "id"]
```

```python
x = [i for i in train.names if i not in ignoreFields]
```

```python
x
```

```python
ml = H2OGradientBoostingEstimator(model_id = "defaults")
```

```python
ml.train(x,y,train, validation_frame = valid)
```

```python
ml.mae(train=True)
```

```python
ml.mae(valid=True)
```

```python
perf = ml.model_performance(test)
```

```python
perf
```

```python
m2 = H2OGradientBoostingEstimator(model_id = "overfit_python")
m2.train(x,y,train, validation_frame = valid)
```

```python
print("Train: %d", ml.mae(train=True),  m2.mae(train=True))
print("Valid: %d", ml.mae(valid=True),  m2.mae(valid=True))
print("Test: %d", perf.mae(),  m2.model_performance(test).mae())
```
